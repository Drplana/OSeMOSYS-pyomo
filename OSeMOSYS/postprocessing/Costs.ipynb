### Capital Cost by Technology

```python
CapitalInvestment = pd.read_csv('../../results/v_CapitalInvestment.csv')
CapitalInvestment = CapitalInvestment[CapitalInvestment['TECHNOLOGY']=='PWRCRUD008']
```

```python
OperatingCost = pd.read_csv('../../results/v_OperatingCost.csv')
OperatingCost = OperatingCost[OperatingCost['TECHNOLOGY']=='PWRCRUD008']
```

```python
EnergyProduction = pd.read_csv('../../results/v_ProductionByTechnologyAnnual.csv')
filtro =  (EnergyProduction['TECHNOLOGY']=='PWRCRUD008')&(EnergyProduction['FUEL']=='ELC01')
EnergyProduction = EnergyProduction[filtro]
```

```python
EnergyProduction
```

```python
mx1 = pd.merge(EnergyProduction, OperatingCost, on = ["REGION", "TECHNOLOGY", "YEAR"],how="outer",
    suffixes=("_energy", "_opcost"))
mx2 = pd.merge(mx1, CapitalInvestment, on = ["REGION", "TECHNOLOGY", "YEAR"],how="outer")
mx2
```
